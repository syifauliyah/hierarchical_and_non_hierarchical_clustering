# Modelling

```python
# Import Beberapa Library
from sklearn import datasets,metrics
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
from sklearn.datasets import load_iris
import numpy as np
from sklearn.metrics import silhouette_score
import pandas as pd
from scipy.spatial import distance
from google.colab import files
from IPython.display import Image
import scipy.cluster.hierarchy as sch
from sklearn.neighbors import NearestNeighbors
```

```python
# Import Data
data = pd.read_csv('data.csv', sep=";")
```

```python
# Delete Kolom
del_cols = ['Kecamatan', 'JumlahDosis1', 'JumlahDosis2', 'JumlahDosis3', 'TotalVaksinDiberikan',
            'LansiaTotalVaksinDiberikan', 'PelayanPublikTotalVaksinDiberikan', 'GotongRoyongTotalVaksinDiberikan',
            'TenagaKesehatanTotalVaksinDiberikan', 'Tahapan3TotalVaksinDiberikan', 'RemajaTotalVaksinDiberikan']
data.drop(columns=del_cols, axis=1, inplace=True)
data.head(5)
```

### K-Means Clustering

Dilakukan pencarian nilai k terlebih dahulu menggunakan metode elbow dan metode silhouette 

```python
# ELBOW METHOD
sse = {}
wcss=[]
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300,
                    n_init = 10, random_state = 0)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
    sse[i] = kmeans.inertia_
    print("For cluster = {}, SSE/WCSS is {}".format(i, sse[i]))

plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
# Silhouette Score
wcss = []

for k in range(2,20):
  kmeans = KMeans(n_clusters= k,max_iter=100).fit(data)
  label = kmeans.labels_
  sil_coeff = silhouette_score(data,label,metric = 'euclidean')
  print('For cluster= {}, Silhouette Coefficient is {}'.format(k,sil_coeff))
print('\n')

cluster_Kmeans = KMeans(n_clusters=3)
model_kmeans = cluster_Kmeans.fit(data)
labels_1 = model_kmeans.labels_
print("Silhouette Score for 3 Cluster")
print(silhouette_score(data,labels_1))
```

Maka Berdasarkan Clustering menggunakan Elboww Method dan Sihouette Score didapatkan K=3 Sehingga digunakan lah k=3 pada k-means clustering ini

```python
# Menjalankan Klastering
cluster_Kmeans = KMeans(n_clusters=3)
model_kmeans = cluster_Kmeans.fit(data)
pred_kmeans = model_kmeans.labels_
print(pred_kmeans)

# Frekuensi Klaster
unique, counts = np.unique(pred_kmeans, return_counts=True)
dict(zip(unique, counts))
```

```python
hasil = {
    'Kecamatan' : df['Kecamatan'],
    'Cluster' : pred_kmeans
}
dfa = pd.DataFrame(hasil)
dfa
```

Hasil menggunakan k-means klustering didapatkan 3 klaster dimana 21 kecamatan termasuk daerah dengan tinkat vaksinasi sedang, 17 kecamatan termasuk daerah dengan tingkat vaksinasi rendah dan 6 kecamatan termasuk daerah dengan tingkat vaksinasi tinggi.

Salah satu contoh daerah dengan tingkat vaksinasi tinggi adalah Kecamatan Duren Sawit, Kecamatan Jaga Karsa salah satu contoh daerah dengan tingkat vaksinasi rendah dan Kecamatan Gambir adalah salah satu daerah dengan tingkat vaksinasi sedang.

```python
data = np.array(data) #that all
plt.scatter(data[pred_kmeans==0,0],data[pred_kmeans==0,1],s=50, c='yellow',label='Sedang')
plt.scatter(data[pred_kmeans==1,0],data[pred_kmeans==1,1],s=50, c='red',label='Rendah')
plt.scatter(data[pred_kmeans==2,0],data[pred_kmeans==2,1],s=50, c='green',label='Tinggi')
plt.legend()
```

Di atas adalah hasil klasterisasi menggunakan k-means dengan Kecamatan dengan tingkat vaksinasi rendah sebanyak 17 kecamatan, tingkat vaksinasi sedang sebanyak 21 kecamatan dan tingkat vaksinasi tinggi sebanyak 6 kecamatan

Sehingga pemerintah harus lebih fokus menggencarkan vaksinasi di daerah dengan tingkat vaksinasi rendah supaya batasan-batasan berkegiatan masyarakat lebih minim dan kegiatan ekonomi masyarakat dapat berjalan lebih baik lagi.
